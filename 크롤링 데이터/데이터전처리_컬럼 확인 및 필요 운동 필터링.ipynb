```python
import pandas as pd
import os

#운동 부위에 따른 운동 부위 전처리
filename = 'Shoulders' #운동 부위 : abs, arms, back, butt, chest, Legs-Thighs, Legs-Calves and Shins, Full-body, Shoulders

file_list = os.listdir('img2/'+filename)
index = []
training_name = []
training_index = []

for i in range(len(file_list)) :
    #각 운동 부위 운동 이미지, 순서 추출
    training_index.append(file_list[i][file_list[i].find('-',len(file_list[i])-7)+1:].split('.')[0])
    training_name.append(file_list[i][:file_list[i].find('-',len(file_list[i])-7)])
    index.append(i+1)


file = pd.DataFrame({'index':index,'training_name':training_name,'training_index':training_index})
```

```python
file
```

```python
#파일 저장
file.to_csv(filename+'_img_info.csv', index=False, encoding='cp949')
```
